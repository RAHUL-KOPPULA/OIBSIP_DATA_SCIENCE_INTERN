```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (14, 6)
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r"C:\Users\RAHUL BABU KOPPULA\OneDrive\Desktop\Oasis_ infobyte_internship\datasets\Unemployment_Rate_upto_11_2020.csv")
df
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.rename(columns={'Estimated Unemployment Rate (%)': 'Unemployment_Rate'}, inplace=True)
```

```python
df
```

```python
print(df.columns.tolist())
```

```python
df = df.sort_values('Date')
df
```

```python
df_avg = df.groupby('Date')['Unemployment_Rate'].mean()
df_avg
```

```python
plt.plot(df_avg.index, df_avg.values, label='Average Unemployment Rate (%)', color='blue')
plt.axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2021-03-01'), color='red', alpha=0.2, label='COVID-19 Peak')
plt.title("Average Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.plot(region_data['Date'], region_data['Unemployment_Rate'], label=region)
plt.title("Unemployment Rate by Region")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()    
```

```python
print("Overall Unemployment Rate Stats:")
print(df['Unemployment_Rate'].describe())
```

```python
covid_period = df[(df['Date'] >= '2020-03-01') & (df['Date'] <= '2021-03-01')]
print("\nCOVID-19 Period Stats:")
print(covid_period['Unemployment_Rate'].describe())
```

```python
df_avg = df.groupby('Date')['Unemployment_Rate'].mean()
df_avg_diff = df_avg.diff()

plt.plot(df_avg.index, df_avg_diff, color='red')
plt.axhline(0, color='black', linestyle='dotted')
plt.title("Monthly Change in Average Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Change (%)")
plt.show()
```
